## Black-Scholes Diffusion Model

https://quant.stackexchange.com/questions/10939/black-scholes-diffusion-well-coded-in-python

```python
mu  = 0
sig = 0.5

def generate(S0, T, nt, sig, mu):
    nt = int(nt)
    T = float(T)
    dt  = T / nt

    St  = [S0] * nt

    t   = np.linspace(0,T,nt)
    dWt = np.random.normal(0, 1, nt)
    dWt[0] = 0.
    Wt  = dWt.cumsum()

    return S0 * np.exp(sig * np.sqrt(float(dt)) * Wt + (mu - sig**2/2.) * t)
```

```python
for i in range(100):
    plt.plot(generate(180., 2., 252*2, sig, mu))
```
